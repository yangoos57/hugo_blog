```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
mat_1= np.array([3,3,1,6]).reshape(2,2)
mat_2= np.array([4,2,1,5]).reshape(2,2)
mat_3= np.array([6,0,2,4]).reshape(2,2)
```

```python
font = {'family': 'serif',
        'color':  'black',
        'weight': 'normal',
        'size': 16,
        }
```

```python
fig, ax = plt.subplots(1,3,figsize=(16,8))
# sns.set(font_scale=1.5)
plt.rc('font', family='Malgun Gothic') # For Windows
plt.subplot(1,3,1)
sns.heatmap(mat_1,annot=True, cmap='PuBu', cbar=False, xticklabels=['Positive', 'Negative'], yticklabels=['Positive', 'Negative'],square=True,fmt='g',annot_kws={'size':18}, ax=ax[0] )
plt.title('\n전담팀장\n',fontdict={'weight' : 'bold'})

plt.subplot(1,3,2)
sns.heatmap(mat_2,annot=True, cmap='PuBu', cbar=False, xticklabels=['Positive', 'Negative'], yticklabels=['Positive', 'Negative'],square=True,fmt='g',annot_kws={'size':18}, ax=ax[1])
plt.title('\n수사부장\n',fontdict={'weight' : 'bold'})

plt.subplot(1,3,3)
sns.heatmap(mat_3,annot=True, cmap='PuBu', cbar=False, xticklabels=['Positive', 'Negative'], yticklabels=['Positive', 'Negative'],square=True,fmt='g',annot_kws={'size':18} , ax=ax[2])
plt.title('\n경찰청장\n',fontdict={'weight' : 'bold'})

fig.text(0.5, 0.1, 'Predict Label', ha='center',fontdict={'fontsize': '20'})
fig.text(0.01, 0.5, 'True Label', va='center', rotation='vertical',fontdict={'fontsize': '20'})
plt.margins(x=1, y=1)
# plt.tight_layout()
```

.1e = scientific notation with 1 decimal point (standard form)

.2f = 2 decimal places

.3g = 3 significant figures

.4% = percentage with 4 decimal places

```python
mat_4 = np.array([1.0,0.5,0.8,0.67,0.75,1]).reshape(3,2)

fig, ax = plt.subplots()

sns.heatmap(mat_4, annot=True, cmap='Greys',cbar=False, yticklabels=['팀장','부장','청장'],xticklabels=['Precision','Recall'],fmt='.0%')

ax.xaxis.set_ticks_position('top')
ax.xaxis.set_label_position('top')
```
